# Seperate labels and target and and train a , test and train model

```python
x = dataset.drop(columns=['R'],axis = 1 )
y = dataset['R']

# x.shape
# y.shape


x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)

print("x train shape is :- ",x_train.shape)
print("x test shape is :- ",x_test.shape)
print("y train shape is :- ",y_train.shape)
print("y test shape is :- ",y_test.shape)
```

```python
model = LogisticRegression()

model.fit(x_train,y_train)

p_pred = model.predict(x_test)
```

# Model Evaluation

```python
accuracy = accuracy_score(y_test,p_pred)

print(f"Accuracy of the model is :- {round(accuracy*100, 2)}%")
```

```python
# Accuracy on training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print(f'Accuracy on training data : {round(training_data_accuracy*100,2)}%')
```
